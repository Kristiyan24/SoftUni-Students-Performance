# Training Data Analyze

```python
import pandas as pd
```

```python
def print_columns(data):
    print(data.columns)

def print_info(data):
    print(data.info())

def print_analyze(data):
    for col in data.columns:
        if col != "student_id":
            items = (data[col].dropna().astype(str).value_counts().sort_index())
            print(f"Column: {col}")
            print("─" * 40)
            for value, count in items.items():
                print(f'==>  "{value}" - {count} count')
            print("\n")
```

```python
training_data = pd.read_csv('../../data/train.csv')
```

```python
print_columns(training_data)
```

```python
print_info(training_data)
```

```python
print_analyze(training_data)
```

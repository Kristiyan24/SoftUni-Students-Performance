## Setup:

```python
import pandas as pd
```

```python
def print_table(data):

    table_styling = [
        ('width', '100%'),
        ('border', '1px solid #DDD'), 
        ('font-family', 'sans-serif')
    ]

    header_styling = [
        # text styling
        ('text-align', 'left'), 
        ("font-size", 'large'),
        ("font-weight", 'bold'),

        # text box coloring
        ('color', '#FFFFFF'),
        ('background-color', '#4CAF50'),
        ('border', '1px solid #DDD'), 
        
        # other stylings
        ('padding', '15px 15px 15px 15px'), 
    ]

    data_styling = [
        # text styling
        ('text-align', 'left'), 
        ("font-weight", 'bold'),
        ("font-size", 'large'),

        # text box coloring
        ('color', '#FFFFFF'),
        ('background-color', '#000000'),
        ('border', '1px solid #DDD'), 
        
        # other stylings
        ('padding', '10px 10px 10px 10px'), 
    ]

    data_hover_styling = [
        ('color', '#000000'),
        ('background-color', '#DDD')
    ]

    result = data.style.set_table_styles([
        # Table styling
        {'selector': 'table', 'props': table_styling},
        # Header styling
        {'selector': 'th', 'props': header_styling},
        # Data styling
        {'selector': 'td', 'props': data_styling},
        # Hover effect
        {'selector': 'tbody tr:hover td', 'props': data_hover_styling}
    ])

    result.hide(axis='index') # Hides the default 0, 1, 2... index

    return result
    
```

## Data:
